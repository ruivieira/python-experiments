# intrplate

```python
import numpy as np
import intrplate

data = np.array([[18, 50e3], [31, 100e3], [47, 200e3], [60, 300e3]])
```

```python
import matplotlib.pyplot as plt

plt.scatter(data[:,0], data[:,1])
plt.show()
```

```python
interpolated = intrplate.basis_spline(data=data, degree=3)
```

```python
plt.scatter(interpolated[:,0], interpolated[:,1], s=2)
plt.scatter(data[:,0], data[:,1], c='red')
plt.show()
```

```python
interpolated_noisy = intrplate.noisy_basis_spline(data=data, degree=3, std=5.0)
```

```python
plt.scatter(interpolated_noisy[:,0], interpolated_noisy[:,1], s=2)
plt.scatter(data[:,0], data[:,1], c='red')
plt.show()
```
